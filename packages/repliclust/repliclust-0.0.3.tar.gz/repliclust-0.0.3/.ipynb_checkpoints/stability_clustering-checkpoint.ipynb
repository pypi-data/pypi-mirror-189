```python
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap

class StabClust():
    """
    Compute stable clustering.
    """
    
    def __init__(self,n_subsamples=100,f=0.5,model=KMeans(),**kwargs):
        """
        Initiate StableCluster object.
        """
        self.X = None
        self.y = None
        self.alpha = None
        self.cluster_membership_probs = None
        self.n_clusters = None
        self.f_stably_clustered = None
        self.n_subsamples = n_subsamples
        self.f = f
        
        # set clustering algorithm parameters
        self.model = model
        for key in kwargs.keys():
            setattr(self.model, key, kwargs[key])
            

    def compute_k(self, X, n_repeat=5):
        """
        Compute the number of clusters.
        """
        n_clusters = np.zeros(n_repeat, dtype='int')
        
        for i in range(len(n_clusters)):
            y_sub, S = self.cluster_a_subsample(X)
            n_clusters[i] = len(np.unique(y_sub))
            
        if len(np.unique(n_clusters)) > 1:
            raise Exception('inferred number of clusters not consistent')
            
        return n_clusters[0]
    
    
    def cluster_a_subsample(self,X):
        """
        Cluster a subsample of the data and return the predicted cluster labels.
        """
        S = np.random.choice(X.shape[0],int(X.shape[0]*self.f),replace=False)
        X_sub = X[S,:]
        y_sub = self.model.fit_predict(X_sub)
        return y_sub, S
        
            
    def cluster_membership(self, X):
        """
        Compute the probabilities that each observation belongs a given cluster.
        """
        n = X.shape[0]
        p = X.shape[1]
        
        if self.n_clusters is None:
            raise Exception('determine the number of clusters first')
        else:
            k = self.n_clusters
        
        obs_in_subsample = np.zeros(n)
        cluster_membership = np.zeros(shape=(n,k))
        
        # cluster the whole dataset
        y_ref = self.model.fit_predict(X)

        for i in range(self.n_subsamples):
            y_sub, S = self.cluster_a_subsample(X)

            # if this is not the first iteration, match cluster indices
            if (i > 0):
                # rows are current cluster indices, cols are existing cluster indices
                cluster_match = np.zeros(shape=(k,k))
                for local_idx, cluster_idx in enumerate(y_sub):
                    #if obs_in_subsample[S[local_idx]] > 0:
                    #    cluster_match[cluster_idx,:] += cluster_membership[S[local_idx],:]
                    cluster_match[cluster_idx,y_ref[S[local_idx]]] += 1

                y_sub = np.argmax(cluster_match,axis=1)[y_sub]

            for local_idx, cluster_idx in enumerate(y_sub):
                cluster_membership[S[local_idx],cluster_idx] += 1
                obs_in_subsample[S[local_idx]] += 1
                
        cluster_membership_probs = cluster_membership/obs_in_subsample[:,np.newaxis]
        
        return cluster_membership_probs
    
    
    def fit(self,X):
        """
        Apply stable clustering to the data.
        """
        self.X = X
        self.n_clusters = self.compute_k(X)
        self.cluster_membership_probs = self.cluster_membership(X)
        
        
    def predict(self,alpha=0.05):
        """
        """
        if self.cluster_membership_probs is None:
            raise Exception('need to fit StabClust to data first, for example by using'+
                            'StabClust.fit')
        # otherwise return the predictions
        y = np.argmax(self.cluster_membership_probs,axis=1)
        y[np.max(self.cluster_membership_probs,axis=1)<1-alpha] = -1
        self.f_stably_clustered = np.mean(y>=0)
        
        # set the predictions as an attribute
        self.y = y
        self.alpha = alpha
        
        return self.y
        
    
    def fit_predict(self,X,alpha=0.05):
        """
        Classify the observations with stable clustering.
        """
        self.fit(X)
        return self.predict(alpha=alpha)
    
    
    def instability(self):
        """
        Compute the instability for this clustering.
        """
        if self.cluster_membership_probs is None:
            raise Exception('need to fit StabClust to data first, for example by using'+
                            'StabClust.fit')
        return np.mean(entropy(self.cluster_membership_probs,axis=1))/np.log(self.n_clusters)
    
    
    def plot(self,plot_unstable=True):
        """
        """
        if self.X is None:
            raise Exception('need to fit StabClust to data first, for example by using'+
                            'StabClust.fit')
        if self.y is None:
            raise Exception('need to predict cluster labels first, for example by using'+
                            'StabClust.predict')
            
        if (self.X.shape[1] > 2):
            X_2d = compute_2d_rep(self.X)
        else:
            X_2d = X
        
        cluster_idx = np.unique(self.y[self.y != -1])
        n_clusters = len(cluster_idx)
        
        plt.figure()
        if plot_unstable:
            plt.scatter(X_2d[self.y==-1,0], X_2d[self.y==-1,1], color='black', label='unstable',alpha=1)
        for i,k in enumerate(cluster_idx):
            plt.scatter(X_2d[self.y==k,0], X[self.y==k,1], color=self.cluster_color(k, n_clusters),
                        label='Cluster ' + str(i+1))
        plt.xlabel('x')
        plt.ylabel('y').set_rotation(0)
        
        plt.title(r'Stable Clustering into $k=$' + str(n_clusters) + r' clusters ($\alpha=$' + str(self.alpha) + ')');
        plt.legend()
        
    
    def compute_2d_rep(self, method='umap'):
        """
        """
        if self.X is None:
            raise Exception('need to fit StabClust to data first, for example by using'+
                            'StabClust.fit')
        
        if (self.X.shape[1] == 2):
            self.X_2d = self.X
        elif (self.X.shape[1] > 2):
            # apply dimension reduction
            pass
        
            
    def cluster_color(self, k, n_clusters, cmap='viridis'):
        """
        Compute color for the k-th cluster.
        """
        return get_cmap(cmap)((k+1)/(n_clusters+1))
```
